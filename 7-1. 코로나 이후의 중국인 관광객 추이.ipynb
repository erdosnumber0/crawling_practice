# 데이터 전처리 과정을 함수로 만들기

1. 불러올 데이터 파악
2. 엑셀 파일 불러오기
3. 데이터 탐색
4. 기준년월 칼럼 추가
5. 국적만 남기기(대륙 제거)
6. 대륙 칼럼 삽입
7. 국적별 관광객 비율(%)
8. 전체 외국인 관광객 대비 국적별 관광객 비율(%)

```python
def create_kto_data(yy,mm):
    # 1.불러올 데이터 파악
    file_path = './dataset_7/kto_{}{}.xlsx'.format(yy,mm)
    
    # 2. 엑셀 파일 불러오기
    df = pd.read_excel(file_path, header=1, skipfooter = 4, uscols = 'A:G') # 위와 동일
    
    # 4. 기준 년월 칼럼 추가
    df['기준년월']= '{}-{}'.format(yy,mm)
    
    # 5. 국적 칼럼에서 대륙 제거하고 국가만 남기기
    ignore_list = ['아시아주', '미주', '구주', '대양주', '아프리카주','기타대륙', '교포소계']
    condition = (df.국적.isin(ignore_list)==False)
    df_country = df[condition].reset_index(drop=True)
    
    # 6. 대륙 칼럼 삽입
    continents = ['아시아'] *25 + ['아메리카'] * 5 +['유럽'] *23 +['오세아니아'] *3 + ['아프리카']*2 +['기타대륙'] +['교포']
    df_country['대륙']=continents
    
    # 7.국적별 관광객비율 칼럼 추가
    df_country['관광객비율(%)'] = round(df_country.관광/df_country.계 * 100, 1)
    
    # 8. 전체 외국인 관광객 대비 국적별 관광객비율 칼럼 추가
    df_country['전체비율(%)']= round(df_country['관광']/t_sum *100, 1)
    
    return(df_country)
    
```

# 여러 엑셀 파일 하나로 합치기

```python
# 2010~2020, 1~12월 , 과정1

for yy in range(2010, 2021):
    for mm in range(1,13):
        mm_str = str(mm).zfill(2) # 괄호 안의 숫자만큼 0을 채워서 자릿수 맞춤
        yymm = '{}{}'.format(yy,mm_str)
        print(yymm)
```

```python
# 과정2 , 데이터프레임, 사용자함수
import time

df = pd.DataFrame()

start = time.time()
for yy in range(2010, 2021):
    for mm in range(1,13):
        temp = create_kto_data(str(yy), str(mm).zfill(2)) # 괄호 안의 숫자만큼 0을 채워서 자릿수 맞춤
        df = df.append(temp, ignore_index=True)
end = time.time() - start
print("used time: %dm%ds" % divmod(end,60))
```

```python
# 과정2 , 데이터프레임, 사용자함수 # 에러 안나게 try/excpt
import time

df = pd.DataFrame()

start = time.time()
for yy in range(2010, 2021):
    for mm in range(1,13):
        try:
            temp = create_kto_data(str(yy), str(mm).zfill(2)) # 괄호 안의 숫자만큼 0을 채워서 자릿수 맞춤
            df = df.append(temp, ignore_index=True)
        except:
            pass
end = time.time() - start
print("used time: %dm%ds" % divmod(end,60))
```
